```python
for col in df.columns:
    print(col,':',df[col].unique())
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('type',axis = 1)
y = df['type']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = KNeighborsClassifier(n_neighbors = 6,p = 4,n_jobs = 1,leaf_size = 40)
```

```python
print(model.get_params())
```

```python
model.fit(x_train,y_train)
```

```python
#checking for the overfitting or underfitting
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
#to predict the model
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
classification_report = classification_report(y_pred,y_test)
print(classification_report)
```

```python
confusion_matrix = confusion_matrix(y_pred,y_test)
print(confusion_matrix)
```

```python
accuracy_score = accuracy_score(y_pred,y_test)
print(accuracy_score)
```

```python
----->What distance metrics can be used in KNN?
Distance metrics tell KNN how to measure “closeness” between points (data instances). Choosing the right one affects the predictions.
*Euclidean Distance
*Manhattan Distance
*Minkowski Distance
*Chebyshev Distance
*Cosine Distance
```
